```python
import pandas as pd
import os

# 加载数据
data = pd.read_csv('H:/project/SVM/dataset/9-PTFE.csv', header=None)

# 假设所有列都是同一种材料，比如 'woolly'
material_type = 'PTFE'

# 创建一个与数据列数相同的列表，填充材料类型
material_labels = [material_type] * data.shape[1]
new_row = pd.DataFrame([material_labels], columns=data.columns)

# 将新行与原始数据框合并
data = pd.concat([new_row, data], ignore_index=True)
print("数据的维数:", data.shape)

# 检查并创建保存目录
save_directory = 'H:/project/SVM/data'
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# 保存文件到指定目录
data.to_csv(os.path.join(save_directory, 'PTFE.csv'), index=False, header=False)

# 显示前几行数据
print("数据的前几行:")
print(data.head())
print("列名:", data.columns)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# 读取CSV文件
# data = pd.read_csv('H:/project/SVM/combined_data/unpre_combined_data.csv', header=0)
data = pd.read_csv('H:/project/SVM/data/A4.csv', header=None)
print("清理前的数据：")
print(data.head())

# 去除缺失值
data.dropna(inplace=True)
some_threshold = 0.02

# 定义一个函数来去除无效数据段和突变部分
def clean_data(column):
    # 从第二行开始去除开头的无效数据段
    start_index = 1  # 从第二行开始
    for i in range(1, len(column)):  # 从索引1开始
        if column[i] > some_threshold:  # some_threshold是你定义的有效数据的阈值
            start_index = i
            break
    
    # 去除结尾的无效数据段和突变部分
    end_index = len(column)
    for i in range(len(column) - 1, -1, -1):
        if column[i] < -0.02:
            end_index = i + 1
            break
    
    # 返回清理后的数据
    return column[start_index:end_index]

# 对每一列应用清理函数
for col in data.columns:
    data[col] = clean_data(data[col])

# 检查并创建保存目录
save_directory = 'H:/project/SVM/combined_data'
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# 保存文件到指定目录，保留header
data.to_csv(os.path.join(save_directory, 'pre_combined_data.csv'), index=False, header=True)

# 检查清理后的数据
print("清理后的数据预览：")
print(data.head())

# 绘制前三列数据，从第二行开始
plt.figure(figsize=(12, 6))

for i in range(3):
    plt.plot(data.iloc[1:, i], label=f'Column {i+1}')  # 使用iloc从第二行开始

plt.title('first 3 columns of the data')
plt.xlabel('time point')
plt.ylabel('measurement value')
plt.legend()
plt.show()
```
